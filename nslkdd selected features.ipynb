### FEATURE SELECTION FROM FEATURE IMPORTANCE OF RANDOM FOREST CLASSIFIER

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
sel = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1))
sel.fit(x_dupli_train, y_dupli_train)
sel.get_support()
```

```python
features = x_dupli_train.columns[sel.get_support()]
len(features)
```

```python
np.mean(sel.estimator_.feature_importances_)
```

```python
sel.estimator_.feature_importances_
```

```python
x_dupli_train_rfc= sel.transform(x_dupli_train)
x_dupli_test_rfc= sel.transform(x_dupli_test)
```

```python
def run_randomforest(x_dupli_train, x_dupli_test, y_dupli_train, true_p):
    clf= RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
    clf.fit(x_dupli_train, y_dupli_train)
    y_pred= clf.predict(x_dupli_test)
    print('accuracy: ', accuracy_score(true_p, y_pred))
```

```python
%%time
run_randomforest(x_dupli_train_rfc, x_dupli_test_rfc, y_dupli_train, true_p)
```

```python
%%time
run_randomforest(x_dupli_train, x_dupli_test, y_dupli_train, true_p)
```

### GRADIENT BOOST

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
sel= RFE(GradientBoostingClassifier(n_estimators=100, random_state=0), n_features_to_select=10)
sel.fit(x_dupli_train, y_dupli_train)
sel.get_support()
```

```python
features= x_dupli_train.columns[sel.get_support()]
features
```

```python
x_dupli_train_gb= sel.transform(x_dupli_train)
x_dupli_test_gb= sel.transform(x_dupli_test)
```

```python
def run_gradientboost(x_dupli_train, x_dupli_test, y_dupli_train, true_p):
    clf= GradientBoostingClassifier(n_estimators=100, random_state=0)
    clf.fit(x_dupli_train, y_dupli_train)
    y_pred= clf.predict(x_dupli_test)
    return('accuracy: ', accuracy_score(true_p, y_pred))
```

```python
run_gradientboost(x_dupli_train_gb, x_dupli_test_gb, y_dupli_train, true_p)
```

```python
for index in range(1,2):
    sel = RFE(GradientBoostingClassifier(n_estimators=100, random_state=0), n_features_to_select=index)
    sel.fit(x_dupli_train, y_dupli_train)
    x_dupli_train_gb= sel.transform(x_dupli_train)
    x_dupli_test_gb= sel.transform(x_dupli_test)
    print('selected feature: ', index)
    run_randomforest(x_dupli_train_gb,  x_dupli_test_gb,y_dupli_train, true_p)
    print()
```
